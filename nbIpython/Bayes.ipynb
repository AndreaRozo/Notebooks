```python
mov = loadtxt('movimiento.dat')
x_obs = mov[:,0]
y_obs = mov[:,1]
```

```python
scatter(x_obs,y_obs)
```

```python
def L(y_obs,y_model):
    chi = sum((y_obs-y_model)**2)
    return (-chi)
```

```python
def model(t, a0, a1, a2):
    return (a0 + a1*t + a2*(t**2))
```

```python
a0_walk = empty((0))
a1_walk = empty((0))
a2_walk = empty((0))

a0_walk = append(a0_walk, random.random())
a1_walk = append(a1_walk, random.random())
a2_walk = append(a2_walk, random.random())
```

```python
n_iter = 20000
for i in range(n_iter):
    a0_prime = np.random.normal(a0_walk[i],0.1)
    a1_prime = np.random.normal(a1_walk[i],0.1)
    a2_prime = np.random.normal(a2_walk[i],0.1)
    
    y_init = model(x_obs, a0_walk[i], a1_walk[i], a2_walk[i])
    y_prime = model(x_obs, a0_prime, a1_prime, a2_prime)
    
    alpha = L(y_obs, y_prime)/L(y_obs, y_init)
    if(alpha <= 1.0):
        a0_walk = append(a0_walk, a0_prime)
        a1_walk = append(a1_walk, a1_prime)
        a2_walk = append(a2_walk, a2_prime)
    else:
        beta = random.random()
        if(beta <= exp(-alpha)):
            a0_walk = append(a0_walk, a0_prime)
            a1_walk = append(a1_walk, a1_prime)
            a2_walk = append(a2_walk, a2_prime)
        else:
            a0_walk = append(a0_walk, a0_walk[i])
            a1_walk = append(a1_walk, a1_walk[i])
            a2_walk = append(a2_walk, a2_walk[i])
```

```python
average_a0 = average(a0_walk)
average_a1 = average(a1_walk)
average_a2 = average(a2_walk)
print average_a0
print average_a1
print average_a2
```

```python
best_y = model(x_obs, average_a0, average_a1, average_a2)
scatter(x_obs,y_obs)
plot(x_obs,best_y)
```
