```python
x1 = linspace(-4.0,4.0,1000)
x2 = linspace(-4.0,4.0,1000)

def function(x1,x2):
    num = -(100*((x2-x1**2)**2) + (1-x1)**2)
    return (exp(num/20))

f = function(x1,x2)
```

```python
x1_walk = empty((0))
x1_0 = 8.0*((random.random())-0.5)
x1_walk = append(x1_walk,x1_0)

x2_walk = empty((0))
x2_0 = 8.0*((random.random())-0.5)
x2_walk = append(x2_walk,x2_0)
```

```python
n_iterations = 20000
for i in range(n_iterations):
    x1_prima = np.random.normal(x1_walk[i],0.1)
    x2_prima = np.random.normal(x2_walk[i],0.1)
    alpha = function(x1_prima,x2_prima)/function(x1_walk[i],x2_walk[i])
    if (alpha >= 1.0):
        x1_walk = append(x1_walk,x1_prima)
        x2_walk = append(x2_walk,x2_prima)
    else:
        beta = random.random()
        if (beta <= alpha):
            x1_walk = append(x1_walk,x1_prima)
            x2_walk = append(x2_walk,x2_prima)
        else:
            x1_walk = append(x1_walk,x1_walk[i])
            x2_walk = append(x2_walk,x2_walk[i])
```
