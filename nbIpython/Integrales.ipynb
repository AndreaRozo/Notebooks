```python
h = 0.0000001
n_points = int(5.0/h)
x = linspace(0,5.0,n_points)
y = x*sin(sqrt(x+5))
integral = sum(y)*h
plot(x,y)
real = -8.0*sin(sqrt(5))+38*sin(sqrt(10))-12*sqrt(5)*cos(sqrt(5))+2*sqrt(10)*cos(sqrt(10))
print 'real = ', real, 'integral = ', integral
```

```python
min_y = 0
max_y = 1

n_random = 100000
random_x = random.rand(n_random) * (max_x - min_x) + min_x
random_y = random.rand(n_random) * (max_y - min_y) + min_y

scatter(random_x, random_y, alpha=0.2)
```

```python
delta = (sqrt(1-random_x**2)) - random_y
below  = where(delta>0.0)
scatter(random_x[below], random_y[below])
```
